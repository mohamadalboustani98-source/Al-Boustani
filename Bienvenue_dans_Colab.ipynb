```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import xarray as xr
import pandas as pd
import glob

def pick_one(pattern):
    matches = sorted(glob.glob(pattern))
    if not matches:
        raise FileNotFoundError(f"Missing file matching: {pattern}\nFiles present: {sorted(glob.glob('*.nc'))}")
    return matches[0]

# Pick your uploaded files (no renaming needed)
f_GHI = pick_one("*GHI*.nc")
f_BHI = pick_one("*BHI*.nc")
f_DHI = pick_one("*DHI*.nc")
f_BNI = pick_one("*BNI*.nc")

print("Using files:")
print("GHI:", f_GHI)
print("BHI:", f_BHI)
print("DHI:", f_DHI)
print("BNI:", f_BNI)

def open_any(nc_path):
    # Try the common engines until one works
    for eng in ["h5netcdf", "netcdf4", None]:
        try:
            if eng is None:
                return xr.open_dataset(nc_path)  # let xarray decide
            return xr.open_dataset(nc_path, engine=eng)
        except Exception as e:
            last_err = e
    raise last_err

# Open datasets
ds_GHI = open_any(f_GHI)
ds_BHI = open_any(f_BHI)
ds_DHI = open_any(f_DHI)
ds_BNI = open_any(f_BNI)

# Convert to DataFrames (keeps lat/lon/time indices)
df_GHI = ds_GHI.to_dataframe()
df_BHI = ds_BHI.to_dataframe()
df_DHI = ds_DHI.to_dataframe()
df_BNI = ds_BNI.to_dataframe()

# Merge into one table
df_irr = pd.concat([df_BHI, df_BNI, df_DHI, df_GHI], axis=1)
df_irr.columns = ["BHI", "BNI", "DHI", "GHI"]

# Save
df_irr.to_csv("output_data_irradiance.csv", index=True)
print("‚úÖ Saved: output_data_irradiance.csv")
```

```
import xarray as xr
import pandas as pd
import glob

def to_series(nc_file, out_name):
    ds = xr.open_dataset(nc_file)
    var = list(ds.data_vars)[0]           # the actual data variable
    s = ds[var].to_series()               # multi-index (time, lat, lon)
    s.name = out_name
    return s

f_GHI = sorted([f for f in glob.glob("*.nc") if "GHI" in f])[0]
f_BHI = sorted([f for f in glob.glob("*.nc") if "BHI" in f])[0]
f_DHI = sorted([f for f in glob.glob("*.nc") if "DHI" in f])[0]
f_BNI = sorted([f for f in glob.glob("*.nc") if "BNI" in f])[0]

s_GHI = to_series(f_GHI, "GHI")
s_BHI = to_series(f_BHI, "BHI")
s_DHI = to_series(f_DHI, "DHI")
s_BNI = to_series(f_BNI, "BNI")

df_irr = pd.concat([s_BHI, s_BNI, s_DHI, s_GHI], axis=1).reset_index()
df_irr.to_csv("output_data_irradiance.csv", index=False)

print("Saved fixed output_data_irradiance.csv")
print(df_irr[["GHI","DHI","BNI","BHI"]].describe())
```

```
import xarray as xr
import pandas as pd
import glob

def series_from_nc(nc_file, name):
    ds = xr.open_dataset(nc_file)
    var = list(ds.data_vars)[0]          # real data variable inside that file
    da = ds[var]

    # Mask fill values if present
    fill = da.attrs.get("_FillValue", None)
    if fill is not None:
        da = da.where(da != fill)

    s = da.to_series()
    s.name = name
    return s

f_GHI = sorted([f for f in glob.glob("*.nc") if "GHI" in f])[0]
f_BHI = sorted([f for f in glob.glob("*.nc") if "BHI" in f])[0]
f_DHI = sorted([f for f in glob.glob("*.nc") if "DHI" in f])[0]
f_BNI = sorted([f for f in glob.glob("*.nc") if "BNI" in f])[0]

s_GHI = series_from_nc(f_GHI, "GHI")
s_BHI = series_from_nc(f_BHI, "BHI")
s_DHI = series_from_nc(f_DHI, "DHI")
s_BNI = series_from_nc(f_BNI, "BNI")

df = pd.concat([s_BHI, s_BNI, s_DHI, s_GHI], axis=1).reset_index()
df.to_csv("output_data_irradiance.csv", index=False)

print("Saved: output_data_irradiance.csv")
print(df[["GHI","DHI","BNI","BHI"]].describe())
```

```
import xarray as xr
import pandas as pd
import glob
import matplotlib.pyplot as plt

# -------- 1) Build output_data_irradiance.csv from the 4 NetCDF files --------
def series_from_nc(nc_file, name):
    ds = xr.open_dataset(nc_file)              # let xarray choose engine
    var = list(ds.data_vars)[0]                # real variable inside the file
    da = ds[var]

    # mask fill values if present
    fill = da.attrs.get("_FillValue", None)
    if fill is not None:
        da = da.where(da != fill)

    s = da.to_series()
    s.name = name
    return s

f_GHI = sorted([f for f in glob.glob("*.nc") if "GHI" in f])[0]
f_BHI = sorted([f for f in glob.glob("*.nc") if "BHI" in f])[0]
f_DHI = sorted([f for f in glob.glob("*.nc") if "DHI" in f])[0]
f_BNI = sorted([f for f in glob.glob("*.nc") if "BNI" in f])[0]

s_GHI = series_from_nc(f_GHI, "GHI")
s_BHI = series_from_nc(f_BHI, "BHI")
s_DHI = series_from_nc(f_DHI, "DHI")
s_BNI = series_from_nc(f_BNI, "BNI")

df_irr = pd.concat([s_GHI, s_BHI, s_BNI, s_DHI], axis=1).reset_index()

# detect the time column name produced by reset_index (usually "time")
time_col = [c for c in df_irr.columns if "time" in c.lower()][0]
df_irr[time_col] = pd.to_datetime(df_irr[time_col])

df_irr.to_csv("output_data_irradiance.csv", index=False)
print("‚úÖ Saved: output_data_irradiance.csv")
print(df_irr.head())

# -------- 2) Plot mean BHI / BNI / DHI / GHI vs time --------
ts = df_irr.groupby(time_col)[["GHI", "BHI", "BNI", "DHI"]].mean()

plt.figure(figsize=(11,5))
for col in ts.columns:
    plt.plot(ts.index, ts[col], label=col)

plt.xlabel("Time")
plt.ylabel("Irradiance (W/m¬≤)")
plt.title("Mean irradiance components over Soria (July 2023)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```
import xarray as xr
import pandas as pd
import glob

def pick(pattern):
    m = sorted(glob.glob(pattern))
    if not m:
        raise FileNotFoundError(f"Missing file: {pattern}. ŸÖŸàÿ¨ŸàÿØ ÿßŸÑŸÖŸÑŸÅÿßÿ™: {sorted(glob.glob('*'))[:20]}")
    return m[0]

f_GHI = pick("*GHI*2023_07*.nc")
f_BHI = pick("*BHI*2023_07*.nc")
f_DHI = pick("*DHI*2023_07*.nc")
f_BNI = pick("*BNI*2023_07*.nc")

print("Using:", f_GHI, f_BHI, f_DHI, f_BNI, sep="\n")

def load_series(path, name):
    ds = xr.open_dataset(path)
    var = list(ds.data_vars)[0]
    s = ds[var].to_series()
    s.name = name
    return s

df_irr = pd.concat([
    load_series(f_GHI, "GHI"),
    load_series(f_BHI, "BHI"),
    load_series(f_DHI, "DHI"),
    load_series(f_BNI, "BNI"),
], axis=1).reset_index()

df_irr.to_csv("output_data_irradiance.csv", index=False)
print("‚úÖ Saved output_data_irradiance.csv")
print(df_irr.head())
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("output_data_irradiance.csv")
df["time"] = pd.to_datetime(df["time"])

vars_irr = ["GHI", "BHI", "DHI", "BNI"]

for var in vars_irr:
    # choose the time where this variable has the highest spatial mean
    best_time = df.groupby("time")[var].mean().idxmax()
    d = df[df["time"] == best_time].copy()

    # build 2D grid (lat x lon)
    grid = (
        d.pivot_table(index="latitude", columns="longitude", values=var, aggfunc="mean")
        .sort_index()
    )

    Z = grid.to_numpy()
    lats = grid.index.to_numpy()
    lons = grid.columns.to_numpy()

    plt.figure(figsize=(6,4))
    im = plt.imshow(
        Z,
        origin="lower",
        extent=[float(lons.min()), float(lons.max()), float(lats.min()), float(lats.max())],
        aspect="auto"
    )
    plt.colorbar(im, label=var)
    plt.scatter(d["longitude"], d["latitude"], s=80)  # show the grid points
    plt.xlabel("Longitude (deg)")
    plt.ylabel("Latitude (deg)")
    plt.title(f"{var} heatmap at {best_time}")
    plt.grid(True, linewidth=0.3)
    plt.tight_layout()
    plt.show()
```

```
import xarray as xr, glob
import numpy as np
import matplotlib.pyplot as plt

def open_da(keyword):
    f = [x for x in glob.glob("v4.6_rev2_*_2023_07.area-subset.44.5.0.39.5.-6.nc") if f"_{keyword}_" in x][0]
    ds = xr.open_dataset(f)
    var = list(ds.data_vars)[0]
    return ds[var], f, var

# Daytime mask from GHI (removes night = 0)
da_GHI, fG, vG = open_da("GHI")
day_mask = da_GHI.mean(["latitude","longitude"]) > 10  # sunlit timesteps

print("Day timesteps kept:", int(day_mask.sum().values), "/", int(da_GHI.sizes["time"]))

def plot_day_mean(keyword, label):
    da, f, var = open_da(keyword)
    da_daymean = da.sel(time=day_mask).mean("time")

    vals = da_daymean.values
    vals = vals[np.isfinite(vals)]
    vmin, vmax = np.percentile(vals, 5), np.percentile(vals, 95)

    plt.figure(figsize=(8,5))
    im = plt.pcolormesh(da_daymean.longitude, da_daymean.latitude, da_daymean,
                        shading="auto", vmin=vmin, vmax=vmax)
    plt.colorbar(im, label=label)
    plt.xlabel("Longitude (deg)")
    plt.ylabel("Latitude (deg)")
    plt.title(f"Daytime mean {label} ‚Äî July 2023 (all-sky)")
    plt.tight_layout()
    plt.show()

plot_day_mean("GHI", "GHI (W/m¬≤)")
plot_day_mean("BHI", "BHI (W/m¬≤)")
plot_day_mean("DHI", "DHI (W/m¬≤)")
plot_day_mean("BNI", "BNI (W/m¬≤)")
```
